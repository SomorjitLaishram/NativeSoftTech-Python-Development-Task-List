```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("/content/diabetes_prediction_dataset.csv")

print(df.head())
```

```python
print(df.info())
```

```python

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
plt.tight_layout()
sns.countplot(x='diabetes', data=df, palette='coolwarm')
plt.title("Diabetes Outcome Distribution")
plt.xlabel("Outcome (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True, color='purple')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='diabetes', y='blood_glucose_level', data=df, palette='magma')
plt.title("Blood Glucose Levels by Diabetes Outcome")
plt.xlabel("Diabetes Outcome")
plt.ylabel("Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['bmi'], bins=20, kde=True, color='blue')
plt.title("BMI Distribution of Patients")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for Glucose vs Insulin
plt.figure(figsize=(8,5))
sns.scatterplot(x='blood_glucose_level', y='bmi', hue='diabetes', data=df, palette='coolwarm')
plt.title("Glucose vs Insulin Levels")
plt.xlabel("Glucose")
plt.ylabel("Insulin")
plt.tight_layout()
plt.show()
```
